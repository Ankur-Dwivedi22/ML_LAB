1. Perform Exploratory Data Analysis (EDA) for Iris Species Dataset provided as csv  
file (you can also download it from: https://www.kaggle.com/datasets/uciml/iris ).  
Your EDA should include the following operations:  

Step 1: Import Libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

Step 2: Load the Dataset

```python
df = pd.read_csv('Iris.csv')
```

Step 3: Perform EDA

a. Show size of the dataset

```python
print("Size of the dataset:", df.shape)
```

b. Show datatype for each column

```python
print("Datatypes:\n", df.dtypes)
```

c. Show distribution of data (use describe() function)

```python
print("Data distribution:\n", df.describe())
```

d. Check if there are any null values

```python
print("Null values:\n", df.isnull().sum())
```

e. Check for duplicates

```python
print("Number of duplicates:", df.duplicated().sum())
```

f. Check the number of instances for each species of flower

```python
print("Number of instances for each species:\n", df['Species'].value_counts())
```

g. Compare sepal length and sepal width

```python
sns.scatterplot(x='SepalLengthCm', y='SepalWidthCm', hue='Species', data=df)
plt.title('Sepal Length vs Sepal Width')
plt.show()
```

h. Compare petal length and petal width

```python
sns.scatterplot(x='PetalLengthCm', y='PetalWidthCm', hue='Species', data=df)
plt.title('Petal Length vs Petal Width')
plt.show()
```

i. Use pairplot to show all comparisons

```python
sns.pairplot(df, hue='Species')
plt.show()
```

j. Use histograms to compare sepal length, sepal width, petal length, and petal width across the species

```python
df.hist(column=['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm'], by='Species', bins=20, figsize=(15, 10))
plt.show()
```

k. Use boxplot to show distribution of data across the species

```python
plt.figure(figsize=(12, 8))
sns.boxplot(data=df.drop(columns=['Id']), orient="h", palette="Set2")
plt.title('Boxplot of Features')
plt.show()
```

l. Use violinplot to show distribution of data across the species

```python
plt.figure(figsize=(12, 8))
sns.violinplot(data=df.drop(columns=['Id']), orient="h", palette="Set2")
plt.title('Violinplot of Features')
plt.show()
```
