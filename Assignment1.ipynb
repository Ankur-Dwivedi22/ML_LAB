1. Create an empty and a full array of size 3x3.


```python
import numpy as np
empty_arr = np.empty((3,3))
full_arr = np.full((3,3) , fill_value=1)
print("Empty array : \n" , empty_arr)
print("Full Array : \n", full_arr)
```

2. Generate an array of 25 random numbers sampled from a standard normal distribution.

```python
arr = np.random.normal(0,1,25)
print("Random array : \n", arr)
```

3. Find the dot product of two arrays.

```python
arr1 = np.array([1,2,3])
arr2 = np.array([4,5,6])
dot_product = np.dot(arr1, arr2)
print("Dot Product:\n", dot_product)
```

4. Sort the given array along the row, along the column, and as a whole.

```python
arr = np.array([[3,1,2],[6,4,5],[7,9,8]])
row = np.sort(arr,axis=1)
col = np.sort(arr,axis=0)
mat = np.sort(arr.flatten())
print("Row sorted: \n", row)
print("Column sorted:\n" , col)
print("Whole array sorted: \n", mat)
```

5. Make a list of 3 numpy arrays, find the mean, and output them as a list.

```python
arr_list = [np.array([1,2,3]) , np.array([4,5,6]) , np.array([7,8,9])]
means = [float(np.mean(arr)) for arr in arr_list]
print("Means of arrays:\n", means)
```

6. Make a numpy array from the string, then split it on spaces.

```python
string_arr = np.array(["Starting with Python"])
split_arr = np.char.split(string_arr)
print("Split array: \n", split_arr)
```

7. 
Read Total profit of all months and show it using a line plot. Total profit 
data provided for each month. Generated line plot must include the following 
properties: 
a) X label name = Month Number 
b) Y label name = Total profit

```python
import matplotlib.pyplot as plt
import csv
import pandas as pd

data = pd.read_csv('company_sales_data.csv')
month_numbers = data['month_number']
total_profits = data['total_profit']


plt.figure(figsize=(8,5))
plt.plot(month_numbers , total_profits, marker='o',color='b', linestyle='-')

plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.title('Monthly Profit Line Plot')
plt.grid(True)
plt.show()
```

8. Get total profit of all months and show line plot with the following Style 
properties. Total profit data provided for each month (same as 1st question). 
Generated line plot must include following Style properties: 
c) Line Style dotted and Line-color should be red
d) Show legend at the lower right location. 
e) X label name = Month Number 
f) Y label name = Sold units number 
g) Add a circle marker. 
h) Line marker color as read
i) Line width should be 3 

 




 


```python
data = pd.read_csv('company_sales_data.csv')

months = data['month_number']
total_units = data['total_units']

plt.figure(figsize=(8,5))
plt.plot(months, total_units, color='red', linestyle='dotted', linewidth=3, marker='o', markerfacecolor='red' , label='Total Units Sold')

plt.xlabel("Moth Number")
plt.ylabel('Sold Units Number')
plt.title('Total Units Sold Over Months')
plt.legend(loc='lower right')
plt.grid(True)
plt.show()
```

9.Read toothpaste sales data of each month and show it using a scaler plot. 
Also, add a grid in the plot, gridline style should be . 

```python
tootpaste_sales = data['toothpaste']
plt.figure(figsize=(8,5))
plt.plot(months,tootpaste_sales, color='blue', marker='o', linestyle='--', label='Toothpaste sales')

plt.xlabel('Moth Number')
plt.ylabel('Sales')
plt.title('Toothpaste Sales Over Months')
plt.legend()
plt.grid(True)
plt.show()
```

10.Read face cream and face wash product sales data and show it using the bar 
chart. The bar chart should display the number of units sold per month for each 
product. Add a separate bar for each product in the same chart. 

```python
facewash_sales = data['facewash']
cream_sales = data['facecream']

bar_width = 0.4
x_indexes = range(len(months))

plt.figure(figsize=(8,5))
plt.bar(x_indexes, facewash_sales, width=bar_width, label='Facewash Sales', color='orange')
plt.bar([x + bar_width for x in x_indexes], cream_sales, width=bar_width, label='Cream Sales', color='green')

plt.xlabel('Month Number')
plt.ylabel('Sales')
plt.title('Facewash and Cream Sales')
plt.xticks(ticks=[x + bar_width / 2 for x in x_indexes], labels=months)
plt.legend()
plt.show()
```
